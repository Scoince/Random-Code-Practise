```python
# STEP 2: Import the library and download the video

import yt_dlp

# --- ⬇️ PASTE YOUR TWITTER/X URL HERE ⬇️ ---
video_url = "https://x.com/tetsuoai/status/1976713518058815877"
# --- ⬆️ (This is just an example link) ⬆️ ---

# Set the options for the download
# We'll just name the output file 'my_video.mp4'
ydl_opts = {
    'outtmpl': 'my_video.mp4',
    'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best'
}

# Try to download the video
try:
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([video_url])
    print("✅ Success! Video downloaded as 'my_video.mp4'")

except Exception as e:
    print(f"❌ Failed to download video. Error: {e}")
```
