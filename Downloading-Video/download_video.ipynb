```python
import yt_dlp
import re
from urllib.parse import urlparse

def detect_platform(url):
    """Automatically detect the platform from the URL"""
    url_lower = url.lower()
    domain = urlparse(url).netloc.lower()

    # Platform detection patterns
    platforms = {
        'youtube': ['youtube.com', 'youtu.be'],
        'twitter/x': ['twitter.com', 'x.com'],
        'instagram': ['instagram.com'],
        'tiktok': ['tiktok.com'],
        'facebook': ['facebook.com', 'fb.watch'],
        'reddit': ['reddit.com', 'redd.it'],
        'vimeo': ['vimeo.com'],
        'twitch': ['twitch.tv']
    }

    for platform, domains in platforms.items():
        if any(d in domain for d in domains):
            return platform

    return 'unknown'

def get_download_options(output_format, filename='downloaded_video'):
    """Generate yt-dlp options based on desired output format"""

    base_opts = {
        'outtmpl': f'{filename}.%(ext)s',
        'quiet': False,
        'no_warnings': False,
    }

    if output_format == 'mp4':
        base_opts.update({
            'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best',
            'merge_output_format': 'mp4'
        })

    elif output_format == 'gif':
        base_opts.update({
            'format': 'bestvideo[ext=mp4]/best[ext=mp4]/best',
            'postprocessors': [{
                'key': 'FFmpegVideoConvertor',
                'preferedformat': 'gif',
            }],
            'outtmpl': f'{filename}.%(ext)s',
            'keepvideo': False,
        })

    elif output_format == 'audio':
        base_opts.update({
            'format': 'bestaudio/best',
            'postprocessors': [{
                'key': 'FFmpegExtractAudio',
                'preferredcodec': 'mp3',
                'preferredquality': '192',
            }],
            'outtmpl': f'{filename}.mp3'
        })

    return base_opts

def check_ffmpeg():
    """Check if FFmpeg is installed"""
    import subprocess
    try:
        subprocess.run(['ffmpeg', '-version'], capture_output=True, check=True)
        return True
    except (subprocess.CalledProcessError, FileNotFoundError):
        return False

def download_video(url, output_format='mp4', filename='my_video'):
    """Main download function"""

    # Check FFmpeg for GIF/audio conversion
    if output_format in ['gif', 'audio']:
        if not check_ffmpeg():
            print("‚ö†Ô∏è  FFmpeg is not installed!")
            print("FFmpeg is required for GIF and audio conversion.")
            print("\nInstall FFmpeg:")
            print("  Windows: choco install ffmpeg")
            print("  Mac: brew install ffmpeg")
            print("  Linux: sudo apt install ffmpeg")
            return

    # Detect platform
    platform = detect_platform(url)
    print(f"üîç Detected platform: {platform.upper()}")
    print(f"üì• Downloading as: {output_format.upper()}")
    print("-" * 50)

    # Get appropriate options
    ydl_opts = get_download_options(output_format, filename)

    # Download
    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(url, download=True)
            title = info.get('title', 'Unknown')
            print("-" * 50)
            print(f"‚úÖ Success! Downloaded: {title}")

            if output_format == 'audio':
                print(f"üìÅ Saved as: {filename}.mp3")
            elif output_format == 'gif':
                print(f"üìÅ Saved as: {filename}.gif")
            else:
                print(f"üìÅ Saved as: {filename}.mp4")

    except Exception as e:
        print(f"‚ùå Download failed!")
        print(f"Error: {e}")
        print("\nTroubleshooting tips:")
        print("- Check if the URL is valid and accessible")
        print("- Some platforms may require authentication")
        print("- Try updating yt-dlp: pip install -U yt-dlp")

# ============================================
# USAGE
# ============================================

# Get user input
video_url = input("üìé Paste the video URL: ").strip()

print("\nChoose output format:")
print("1. MP4 (video)")
print("2. GIF (animated)")
print("3. Audio only (MP3)")

format_choice = input("\nEnter choice (1/2/3): ").strip()

# Map choice to format
format_map = {
    '1': 'mp4',
    '2': 'gif',
    '3': 'audio'
}

output_format = format_map.get(format_choice, 'mp4')

# Optional: custom filename
custom_name = input("\nüìù Enter filename (press Enter for default): ").strip()
filename = custom_name if custom_name else 'my_video'

print("\n" + "=" * 50)
print("STARTING DOWNLOAD")
print("=" * 50 + "\n")

# Download the video
download_video(video_url, output_format, filename)
```
